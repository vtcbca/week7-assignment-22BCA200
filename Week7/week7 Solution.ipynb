```python
import csv
```

```python
import pandas as pd
```

```python
field=["Prod_No","Prod_Name","Jan","Feb","Mar","Apr","May","Jun"]
```


1. Add 12 Records. Take input from user.

```python
data=[]
for i in range(12):
    Prod_No=int(input("Enter product no:"))
    Prod_Name=input("Enter product name:")
    Jan=int(input("Enter jan sell:"))
    Feb=int(input("Enter feb sell:"))
    Mar=int(input("Enter march sell:"))
    Apr=int(input("Enter april sell:"))
    May=int(input("Enter may sell:"))
    Jun=int(input("Enter jun sell:"))
    
    data.append([Prod_No,Prod_Name,Jan,Feb,Mar,Apr,May,Jun])
    
```

```python
with open("C:\\henil\\python\\product.csv","w",newline="") as f:
    obj=csv.writer(f)
    obj.writerow(field)
    obj.writerows(data)
    
    
```
